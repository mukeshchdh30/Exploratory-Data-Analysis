## Cleaning The Data

```python
ipl_df.date
```

```python
ipl_df['date'] = pd.to_datetime(ipl_df.date)
```

```python
ipl_df['year'] = pd.DatetimeIndex(ipl_df.date).year
ipl_df['month'] = pd.DatetimeIndex(ipl_df.date).month
ipl_df['day'] = pd.DatetimeIndex(ipl_df.date).day
ipl_df['weekday'] =pd.DatetimeIndex(ipl_df.date).weekday
```

```python
ipl_df.head()
```

* As you can see 4 columns has been increased year, month, day, weekday

```python
#rename some stadium with their present name

ipl_df.venue.replace({'Feroz Shah Kotla Ground':'Feroz Shah Kotla',
                    'M Chinnaswamy Stadium':'M. Chinnaswamy Stadium',
                    'MA Chidambaram Stadium, Chepauk':'M.A. Chidambaram Stadium',
                     'M. A. Chidambaram Stadium':'M.A. Chidambaram Stadium',
                     'Punjab Cricket Association IS Bindra Stadium, Mohali':'Punjab Cricket Association Stadium',
                     'Punjab Cricket Association Stadium, Mohali':'Punjab Cricket Association Stadium',
                     'IS Bindra Stadium':'Punjab Cricket Association Stadium',
                    'Rajiv Gandhi International Stadium, Uppal':'Rajiv Gandhi International Stadium',
                    'Rajiv Gandhi Intl. Cricket Stadium':'Rajiv Gandhi International Stadium'},regex=True,inplace=True)
```

```python
ipl_df.venue
```

```python
#Drop Unwanted columns like umpire3

ipl_df.drop('umpire3', axis = 1, inplace=True)
```

```python
#Here we convert dl_applied column from int64 to bool data type because it consists only 0's and 1's
ipl_df.dl_applied = ipl_df.dl_applied.astype('bool')
```

```python
ipl_df.head(10)
```

## Data Analysis & Visualisation

```python
#Getting the top 5 players with most man of the match awards
ipl_df['player_of_match'].value_counts()[0:5]
```

```python
plt.figure(figsize=(10,8))
plt.bar(list(ipl_df['player_of_match'].value_counts()[0:5].keys()),list(ipl_df['player_of_match'].value_counts()[0:5]))
plt.show()
```

```python
#Finding out the number of toss wins w.r.t each team
ipl_df['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=ipl_df[ipl_df['win_by_runs']!=0]
```

```python
#Looking at the head
batting_first.head()
```

```python
#Making a histogram 
plt.figure(figsize=(8,8))
plt.hist(batting_first['win_by_runs'])
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["green","cyan","red"])
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(8,8))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

 ## Q4: What Team Choses the most After Winning Toss?

```python
team_toss=ipl_df.groupby(['team1','toss_decision']).apply(lambda x:x['toss_decision'].count()).reset_index(name='Toss Decision Counts')
heatmap2_data = pd.pivot_table(team_toss, values='Toss Decision Counts', 
                     index=['team1'], 
                     columns='toss_decision')
heatmap2_data
```

```python
team_toss=ipl_df.groupby(['team1','toss_decision']).apply(lambda x:x['toss_decision'].count()).reset_index(name='Toss Decision Counts')
heatmap2_data = pd.pivot_table(team_toss, values='Toss Decision Counts', 
                     index=['team1'], 
                     columns='toss_decision')

heatmap2_data
fig = plt.figure()
fig, ax = plt.subplots(1,1, figsize=(5,15))
g=sns.heatmap(heatmap2_data,annot=True, cmap="Blues",fmt='g')
g.xaxis.set_ticks_position("top")
ax.set_title('The Toss Decisions taken by teams Heatmap-in IPL',size=20);
```
